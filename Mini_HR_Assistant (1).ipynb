```python
import fitz  # PyMuPDF
import google.generativeai as genai
from google.colab import files

# 1. Upload resume
uploaded = files.upload()
resume_path = list(uploaded.keys())[0]

# 2. Extract text from PDF
def extract_text_from_pdf(path):
    doc = fitz.open(path)
    return "".join([page.get_text() for page in doc])

# 3. Configure Gemini (use your working API key)
# 3. Configure Gemini (use secure input for API key)
import getpass
genai.configure(api_key=getpass.getpass("üîê Enter your API key: "))
model = genai.GenerativeModel(model_name="gemini-1.5-flash")

# 4. Resume Analyzer with ATS
def analyze_resume(resume_text, job_role):
    prompt = f"""
You are an experienced ATS system and HR Expert.

Analyze the following resume for the job role of **{job_role}**.

Resume:
{resume_text}

Give your output in this format using Times New Roman style (Markdown-friendly):

## üéØ Resume Analysis: {job_role}

**1. Matching Skills:**
[List of matching skills]

**2. Missing Skills:**
[List of missing or weak areas]

**3. Fit Percentage:**
Score in percentage

**4. ATS Score:**
Score out of 100

**5. Suggestions:**
Short improvements as bullet points
"""
    response = model.generate_content(prompt)
    return response.text

# 5. Run the analyzer
resume_text = extract_text_from_pdf(resume_path)
job_role = input("Enter job role : ")
result = analyze_resume(resume_text, job_role)

# 6. Print result without error
from IPython.display import Markdown
print("\nüìä AI Resume Analysis:\n")
Markdown(result)
```
